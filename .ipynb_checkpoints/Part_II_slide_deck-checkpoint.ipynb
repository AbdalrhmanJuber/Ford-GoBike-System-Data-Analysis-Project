# Part II: Explanatory Data Analysis - Ford GoBike System
## February 2019 Usage Patterns

This presentation summarizes key findings from the exploratory analysis of the Ford GoBike System data for February 2019. The analysis uncovers patterns in how the bike-sharing system is used across different user types, times, and locations in the San Francisco Bay Area.

```python
# Import required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from math import radians, cos, sin, asin, sqrt
import matplotlib.ticker as ticker

# Set visualization styles
plt.style.use('seaborn-v0_8-whitegrid')
sns.set_palette("deep")
plt.rcParams['figure.figsize'] = [12, 8]
plt.rcParams['font.size'] = 14
plt.rcParams['axes.labelsize'] = 16
plt.rcParams['axes.titlesize'] = 20
plt.rcParams['xtick.labelsize'] = 12
plt.rcParams['ytick.labelsize'] = 12

%matplotlib inline
```

## Dataset Overview

The Ford GoBike System dataset contains trip data from the bike-sharing service in the San Francisco Bay Area for February 2019. Each record represents a single trip and includes information such as:

- Trip duration
- Start and end stations (location and IDs)
- Start and end times
- User type (Subscriber or Customer)
- User demographics (birth year, gender)
- Bike ID and usage program information

After data cleaning and filtering, we analyzed 182,995 bike trips from this period.

```python
# Load the dataset and perform the same cleaning as in the exploratory notebook
df = pd.read_csv('201902-fordgobike-tripdata.csv')

# Convert start_time and end_time to datetime
df['start_time'] = pd.to_datetime(df['start_time'])
df['end_time'] = pd.to_datetime(df['end_time'])

# Create new features
df['trip_day_of_week'] = df['start_time'].dt.day_name()
df['trip_hour'] = df['start_time'].dt.hour
df['trip_duration_min'] = df['duration_sec'] / 60

# Define haversine function for distance calculation
def haversine(lon1, lat1, lon2, lat2):
    # convert decimal degrees to radians 
    lon1, lat1, lon2, lat2 = map(radians, [lon1, lat1, lon2, lat2])
    # haversine formula 
    dlon = lon2 - lon1 
    dlat = lat2 - lat1 
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * asin(sqrt(a)) 
    r = 6371 # Radius of earth in kilometers
    return c * r

# Calculate trip distance
df['trip_distance_km'] = df.apply(lambda row: haversine(
    row['start_station_longitude'], 
    row['start_station_latitude'],
    row['end_station_longitude'], 
    row['end_station_latitude']), axis=1)

# Calculate user age
df['member_birth_year'] = pd.to_numeric(df['member_birth_year'], errors='coerce')
current_year = 2019  # Data is from February 2019
df['member_age'] = current_year - df['member_birth_year']

# Filter out extreme values
df_filtered = df[(df['trip_duration_min'] <= 180) & (df['trip_duration_min'] >= 1)]

# Order days of week correctly
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
df_filtered['trip_day_of_week'] = pd.Categorical(df_filtered['trip_day_of_week'], categories=day_order, ordered=True)

# Display basic statistics
print(f"Total number of trips after filtering: {len(df_filtered):,}")
print(f"Number of unique stations: {df_filtered['start_station_id'].nunique()}")
print(f"Date range: {df_filtered['start_time'].dt.date.min()} to {df_filtered['start_time'].dt.date.max()}")
```

## Key Finding 1: User Types Have Distinct Usage Patterns

The Ford GoBike system has two main user types:
- **Subscribers**: Regular users with membership (89.3%)
- **Customers**: Casual or one-time users (10.7%)

These user groups exhibit significantly different usage patterns in terms of trip duration, time of use, and purpose.

```python
# Create a figure with two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 8))

# Plot 1: User Type Distribution
user_counts = df_filtered['user_type'].value_counts()
ax1.pie(user_counts, labels=user_counts.index, autopct='%1.1f%%', 
        colors=['#3274A1', '#E1812C'], explode=(0.05, 0), 
        textprops={'fontsize': 14}, shadow=True)
ax1.set_title('Distribution of User Types', fontsize=18, pad=20)

# Plot 2: Trip Duration by User Type
sns.boxplot(data=df_filtered, x='user_type', y='trip_duration_min', ax=ax2,
           palette=['#3274A1', '#E1812C'])
ax2.set_title('Trip Duration by User Type', fontsize=18)
ax2.set_xlabel('User Type', fontsize=16)
ax2.set_ylabel('Trip Duration (minutes)', fontsize=16)
ax2.set_ylim(0, 30)  # Focus on the main distribution
ax2.text(0, 24, f"Median: {df_filtered[df_filtered['user_type']=='Subscriber']['trip_duration_min'].median():.1f} min", 
         fontsize=14, ha='center', bbox=dict(facecolor='white', alpha=0.5))
ax2.text(1, 24, f"Median: {df_filtered[df_filtered['user_type']=='Customer']['trip_duration_min'].median():.1f} min", 
         fontsize=14, ha='center', bbox=dict(facecolor='white', alpha=0.5))

plt.tight_layout()
plt.show()

print("Key Observations:")
print("- Subscribers make up the vast majority of users (89.3%)")
print("- Customers take significantly longer trips on average (median: 13.5 vs. 8.5 minutes)")
print("- This suggests different usage purposes: commuting for Subscribers vs. leisure for Customers")
```

## Key Finding 2: Temporal Usage Patterns Reveal Commuting Behavior

The bike-sharing system shows clear commuting patterns on weekdays, with distinct morning and evening peaks, while weekend usage is more evenly distributed throughout the day.

```python
# Create weekday vs weekend column
df_filtered['is_weekend'] = df_filtered['trip_day_of_week'].isin(['Saturday', 'Sunday'])

# Count trips by hour for weekdays and weekends
weekday_trips = df_filtered[~df_filtered['is_weekend']].groupby('trip_hour').size()
weekend_trips = df_filtered[df_filtered['is_weekend']].groupby('trip_hour').size()

# Normalize by dividing by number of days
num_weekdays = df_filtered[~df_filtered['is_weekend']]['start_time'].dt.date.nunique()
num_weekends = df_filtered[df_filtered['is_weekend']]['start_time'].dt.date.nunique()
weekday_trips = weekday_trips / num_weekdays
weekend_trips = weekend_trips / num_weekends

# Plot
plt.figure(figsize=(14, 8))

plt.plot(weekday_trips.index, weekday_trips.values, marker='o', linewidth=3, 
         label='Weekdays (Mon-Fri)', color='#3274A1')
plt.plot(weekend_trips.index, weekend_trips.values, marker='o', linewidth=3, 
         label='Weekends (Sat-Sun)', color='#E1812C')

plt.title('Average Number of Trips by Hour: Weekdays vs. Weekends', fontsize=20)
plt.xlabel('Hour of Day (24-hour format)', fontsize=16)
plt.ylabel('Average Number of Trips', fontsize=16)
plt.xticks(range(0, 24))
plt.grid(True, alpha=0.3)

# Add annotations for key points
plt.annotate('Morning\ncommute\npeak', xy=(8, weekday_trips[8]), xytext=(8, weekday_trips[8]+1000),
            arrowprops=dict(facecolor='black', shrink=0.05, width=1.5), ha='center')
plt.annotate('Evening\ncommute\npeak', xy=(17, weekday_trips[17]), xytext=(17, weekday_trips[17]+1000),
            arrowprops=dict(facecolor='black', shrink=0.05, width=1.5), ha='center')
plt.annotate('Weekend\nleisure\nactivity', xy=(14, weekend_trips[14]), xytext=(14, weekend_trips[14]+1000),
            arrowprops=dict(facecolor='black', shrink=0.05, width=1.5), ha='center')

plt.legend(fontsize=14)
plt.tight_layout()
plt.show()

print("Key Observations:")
print("- Weekdays show two distinct commute peaks at 8-9 AM and 5-6 PM")
print("- Weekend usage is more evenly distributed, with a gradual rise to afternoon peak")
print("- Weekday morning peak is sharper than evening peak, suggesting more synchronized start times")
print("- Night hours (10 PM - 5 AM) show minimal usage on both weekdays and weekends")
```

## Key Finding 3: User Type, Time, and Purpose are Interconnected

Combining user type and temporal patterns reveals how different users utilize the system for different purposes throughout the week.

```python
# Create hour groups for easier visualization
df_filtered['hour_category'] = pd.cut(df_filtered['trip_hour'], 
                                     bins=[0, 6, 10, 15, 19, 24],
                                     labels=['Night (12AM-6AM)', 'Morning (6AM-10AM)', 
                                             'Midday (10AM-3PM)', 'Evening (3PM-7PM)', 
                                             'Night (7PM-12AM)'])

# Prepare data for plotting
user_hour_day = pd.crosstab(index=[df_filtered['user_type'], df_filtered['hour_category']],
                            columns=df_filtered['trip_day_of_week'])

# Reshape for easier plotting
user_hour_day_melt = user_hour_day.stack().reset_index()
user_hour_day_melt.columns = ['user_type', 'hour_category', 'day_of_week', 'count']

# Calculate the total for each group for percentage
user_hour_day_melt['total'] = user_hour_day_melt.groupby(['user_type', 'day_of_week'])['count'].transform('sum')
user_hour_day_melt['percentage'] = user_hour_day_melt['count'] / user_hour_day_melt['total'] * 100

# Set up the figure grid
fig, axes = plt.subplots(2, 1, figsize=(16, 14), sharex=True)

# Subscriber plot
sns.barplot(data=user_hour_day_melt[user_hour_day_melt['user_type'] == 'Subscriber'], 
            x='day_of_week', y='percentage', hue='hour_category', ax=axes[0],
            palette=['#2D3F51', '#3274A1', '#7DADD9', '#E1812C', '#69525E'])
axes[0].set_title('Subscribers: Trip Distribution by Time of Day and Day of Week', fontsize=18)
axes[0].set_ylabel('Percentage of Daily Trips', fontsize=16)
axes[0].legend(title='Time of Day', fontsize=12, title_fontsize=14)
axes[0].grid(axis='y', alpha=0.3)
axes[0].set_xlabel('')

# Customer plot  
sns.barplot(data=user_hour_day_melt[user_hour_day_melt['user_type'] == 'Customer'], 
            x='day_of_week', y='percentage', hue='hour_category', ax=axes[1],
            palette=['#2D3F51', '#3274A1', '#7DADD9', '#E1812C', '#69525E'])
axes[1].set_title('Customers: Trip Distribution by Time of Day and Day of Week', fontsize=18)
axes[1].set_ylabel('Percentage of Daily Trips', fontsize=16)
axes[1].set_xlabel('Day of Week', fontsize=16)
axes[1].legend(title='Time of Day', fontsize=12, title_fontsize=14)
axes[1].grid(axis='y', alpha=0.3)

# Highlight key differences
axes[0].annotate('Commute peaks', xy=(4, 42), xytext=(5, 55),
                arrowprops=dict(facecolor='black', shrink=0.05, width=1.5))
axes[1].annotate('More midday usage', xy=(5, 52), xytext=(3, 65),
                arrowprops=dict(facecolor='black', shrink=0.05, width=1.5))

plt.tight_layout()
plt.show()

print("Key Observations:")
print("- Subscribers show strong commuting patterns on weekdays with morning and evening peaks")
print("- Customers use the system more evenly throughout the day, with highest usage during midday")
print("- Weekend patterns differ significantly from weekdays for both user types")
print("- Subscribers' weekend usage shifts from commuting to leisure (more midday trips)")
print("- This confirms different usage purposes: Subscribers use bikes primarily for commuting on weekdays")
print("  and recreation on weekends, while Customers consistently use bikes more for leisure trips")
```

## Key Finding 4: Trip Duration and Distance Vary by User Type and Purpose

Analyzing trip characteristics reveals further differences between user types and provides insights into how the bike-sharing system is used.

```python
# Create figure with two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 8))

# Plot 1: Distance distribution by user type
sns.histplot(data=df_filtered, x='trip_distance_km', hue='user_type', bins=30,
             kde=True, element='step', common_norm=False, stat='density',
             palette=['#3274A1', '#E1812C'], ax=ax1)
ax1.set_title('Trip Distance Distribution by User Type', fontsize=18)
ax1.set_xlabel('Trip Distance (km)', fontsize=16)
ax1.set_ylabel('Density', fontsize=16)
ax1.grid(True, alpha=0.3)
ax1.set_xlim(0, 8)

# Calculate and display median values
subscriber_median = df_filtered[df_filtered['user_type'] == 'Subscriber']['trip_distance_km'].median()
customer_median = df_filtered[df_filtered['user_type'] == 'Customer']['trip_distance_km'].median()
ax1.axvline(subscriber_median, color='#3274A1', linestyle='--', linewidth=2)
ax1.axvline(customer_median, color='#E1812C', linestyle='--', linewidth=2)
ax1.text(subscriber_median + 0.1, 0.6, f'Median: {subscriber_median:.2f} km', 
         color='#3274A1', fontsize=13, bbox=dict(facecolor='white', alpha=0.7))
ax1.text(customer_median + 0.1, 0.5, f'Median: {customer_median:.2f} km', 
         color='#E1812C', fontsize=13, bbox=dict(facecolor='white', alpha=0.7))

# Plot 2: Scatter plot of duration vs distance with user type
sample_size = 10000  # Sample for better visualization
sample_df = df_filtered.sample(sample_size, random_state=42)
sns.scatterplot(data=sample_df, x='trip_distance_km', y='trip_duration_min', 
                hue='user_type', alpha=0.5, palette=['#3274A1', '#E1812C'], ax=ax2)

# Add trend lines for each user type
for user, color in zip(['Subscriber', 'Customer'], ['#3274A1', '#E1812C']):
    user_data = sample_df[sample_df['user_type'] == user]
    if len(user_data) > 0:
        x = user_data['trip_distance_km']
        y = user_data['trip_duration_min']
        z = np.polyfit(x, y, 1)
        p = np.poly1d(z)
        ax2.plot(np.linspace(0, 8, 100), p(np.linspace(0, 8, 100)), 
                 color=color, linestyle='--', linewidth=3)
        
        # Calculate and display average speed
        avg_speed_kmh = (60 * x / y).median()
        ax2.text(6, p(6) + 5, f'{user}: {avg_speed_kmh:.1f} km/h', 
                 color=color, fontsize=13, bbox=dict(facecolor='white', alpha=0.7))

ax2.set_title('Trip Duration vs Distance by User Type', fontsize=18)
ax2.set_xlabel('Trip Distance (km)', fontsize=16)
ax2.set_ylabel('Trip Duration (minutes)', fontsize=16)
ax2.grid(True, alpha=0.3)
ax2.set_xlim(0, 8)
ax2.set_ylim(0, 60)

plt.tight_layout()
plt.show()

print("Key Observations:")
print("- Most trips are short-distance, with median distances of 1.35 km for Subscribers and 1.67 km for Customers")
print("- Customers not only take longer trips in terms of duration but also slightly longer in terms of distance")
print("- Subscribers tend to ride at higher average speeds, suggesting more direct, purpose-driven trips")
print("- Customer trips show more variability in duration for similar distances, indicating more leisurely pace")
print("- The positive correlation between distance and duration is stronger for Subscribers, suggesting more consistent riding behavior")
```

## Key Finding 5: Popular Stations Serve Different User Needs

The most popular stations in the system show different usage patterns based on their location and the populations they serve.

```python
# Find the top 10 stations
top_stations = df_filtered['start_station_name'].value_counts().head(10).index.tolist()
top_stations_df = df_filtered[df_filtered['start_station_name'].isin(top_stations)]

# Calculate user type proportions for each top station
station_user_props = pd.crosstab(top_stations_df['start_station_name'], top_stations_df['user_type'], normalize='index') * 100

# Sort stations by proportion of subscribers
station_user_props = station_user_props.sort_values(by='Subscriber', ascending=False)

# Create figure
plt.figure(figsize=(14, 10))

# Create the stacked bar chart
station_user_props.plot(kind='barh', stacked=True, color=['#3274A1', '#E1812C'], 
                       figsize=(14, 10), width=0.7)

plt.title('Top 10 Most Popular Stations: User Type Distribution', fontsize=20)
plt.xlabel('Percentage of Trips', fontsize=16)
plt.ylabel('Station Name', fontsize=16)
plt.grid(axis='x', alpha=0.3)
plt.legend(title='User Type', fontsize=14, title_fontsize=16)

# Add percentage labels on bars
for i, (idx, row) in enumerate(station_user_props.iterrows()):
    sub_pct = row['Subscriber']
    cust_pct = row['Customer']
    if sub_pct > 15:  # Only add label if bar is wide enough
        plt.text(sub_pct/2, i, f"{sub_pct:.1f}%", 
                 ha='center', va='center', color='white', fontsize=12, fontweight='bold')
    if cust_pct > 15:
        plt.text(sub_pct + cust_pct/2, i, f"{cust_pct:.1f}%", 
                 ha='center', va='center', color='white', fontsize=12, fontweight='bold')

# Add annotations about station types
plt.annotate('Commuter hubs: \nTransit connections, \nbusiness districts', 
             xy=(95, 2), xytext=(80, 2.5), 
             arrowprops=dict(facecolor='black', shrink=0.05, width=1.5),
             bbox=dict(boxstyle="round,pad=0.5", fc="white", alpha=0.9),
             fontsize=12)

plt.annotate('Mixed-use areas: \nTourism, shopping, \nand commuting', 
             xy=(75, 8), xytext=(60, 9), 
             arrowprops=dict(facecolor='black', shrink=0.05, width=1.5),
             bbox=dict(boxstyle="round,pad=0.5", fc="white", alpha=0.9),
             fontsize=12)

plt.tight_layout()
plt.show()

print("Key Observations:")
print("- Stations near transit hubs and business districts (like BART stations) serve primarily Subscribers (>90%)")
print("- Stations in tourist and mixed-use areas show a higher proportion of Customer users (15-30%)")
print("- This pattern further confirms the primary use cases: commuting for Subscribers and leisure for Customers")
print("- Understanding station usage patterns helps optimize bike distribution and infrastructure planning")
```

## Conclusions

Our analysis of the Ford GoBike System data reveals several important insights:

1. **Dual Purpose System**: The bike-sharing system serves two distinct purposes:
   - A **commuting tool** for regular Subscribers (89% of users)
   - A **leisure activity** for one-time Customers (11% of users)

2. **Temporal Patterns**: Clear usage patterns emerge across different times:
   - **Weekday commute peaks** in the morning (8-9 AM) and evening (5-6 PM)
   - More evenly distributed **weekend leisure usage**

3. **Trip Characteristics**: User types show different behaviors:
   - Subscribers take **shorter, faster, more direct trips** (median: 8.5 min, 1.35 km)
   - Customers take **longer, more leisurely trips** (median: 13.5 min, 1.67 km)

4. **Station Popularity**: Station usage varies based on:
   - **Location** (business district, residential area, tourist attraction)
   - **Day of week** and **time of day**
   - **User demographic** served (commuters vs. tourists)

These findings could help inform system optimization, such as bike rebalancing strategies, station capacity planning, and targeted marketing campaigns to different user segments.
